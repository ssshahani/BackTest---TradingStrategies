### Initialize & Handle Data Function which would be executed every calendar trading day. 

```python
def initialize(context):
    pass

def handle_data(context, data):
    order(symbol('AAPL'), 10)
    record(AAPL=data.current(symbol('AAPL'), 'price'))
```

#### Command to run the algorithm/trading strategy/backtest with parameters
#### Results are retrieved from pickled response.

```python
results = %zipline --bundle quantopian-quandl --start 2017-1-1 --end 2017-12-31 -o returns.pickle
```

### Plot Daily Portfolio Value and Stock Value

```python
import matplotlib.pyplot as plt
ax1 = plt.subplot(211)
results.portfolio_value.plot(ax=ax1)
ax1.set_ylabel('Portfolio Value (USD)')

ax2 = plt.subplot(212)
results.AAPL.plot(ax=ax2)
ax2.set_ylabel('AAPL Stock Price')

plt.tight_layout()
```
